```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.animation as animation

rc('animation', html='jshtml')

%config InlineBackend.figure_format = 'retina'
```

```python
seed = np.random.SeedSequence().entropy
print(seed)
rng = np.random.default_rng(seed=seed)
```

```python
dirichlet_parameters = (.1, .1, .1)

num_of_frames = 100

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(projection='3d')

samples = rng.dirichlet(dirichlet_parameters, (num_of_frames, 1))

def update(t):
  ax.set_xlim(0, 1)
  ax.set_ylim(0, 1)
  ax.set_zlim(0, 1)
  ax.scatter(xs=samples[t,:,0], ys=samples[t,:,1], zs=samples[t,:,2], color="blue")

ax.view_init(elev=20., azim=30)

ani = animation.FuncAnimation(fig, update, num_of_frames, interval=100)
```
